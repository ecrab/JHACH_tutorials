```python
np.random.seed(9) # For reproducibility with examples

with open("sleep_study.csv", "w") as f:

    f.write(f"id,sleep,study\n")
    for i in range(0,1000):

        # Create a random "name" for the subject
        id = 100 +i # This does not matter one bit

        # Coin flip to see if this student will be someone who sleeps a lot
        sleeper = np.random.binomial(1,0.5)
        if sleeper:
            sleep_base = 10
            sleep_noise = np.random.normal(0,1)
        else:
            sleep_base = 3
            sleep_noise = np.random.normal(0,0.75)
        # some noise to add onto the base line sleep amount for high/low sleepers

        sleep = sleep_base + sleep_noise
        if sleep < 0:
            sleep = 0



        # Coin flip to see if this student will be someone who sleeps a lot
        studier = np.random.binomial(1,0.5)

        if studier:
            study_base = 4
            study_noise = np.random.normal(0,0.5)
        else:
            study_base = 1
            study_noise = np.random.normal(0,0.5)

        study = study_base + study_noise
        if study < 0:
            study = 0

        f.write(f"{id},{sleep:0.2f},{study:0.2f}\n")

    f.close()


sleep_study_data = pd.read_csv("sleep_study.csv")
sleep_study_data
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

sleep_study_data = pd.read_csv("sleep_study.csv")

# Use your dataframe columns
X = sleep_study_data[["sleep", "study"]]

inertia = []
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=1602)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # sum of squared distances

# Plot
plt.plot(k_values, inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.title("Elbow Plot")
plt.show()
```

```python
from sklearn.cluster import KMeans

sleep_study_data = pd.read_csv("sleep_study.csv")

# Use your columns
X = sleep_study_data[["sleep", "study"]]

# Choose k (from your elbow plot)
kmeans = KMeans(n_clusters=2, random_state=648)

# Fit model and assign clusters
sleep_study_data["cluster"] = kmeans.fit_predict(X)



plt.scatter(sleep_study_data["sleep"], sleep_study_data["study"], c=sleep_study_data["cluster"], s=5, alpha=0.3)
plt.xlabel("Sleep")
plt.ylabel("Study")
plt.title("K-Means Clustering")
plt.show()
```
